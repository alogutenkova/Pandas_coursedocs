## This Module's Dataset + Memory Optimization

```python
import pandas as pd
```

```python
df = pd.read_csv("employees.csv")
df.head(3)
```

```python
df.info()
```

```python
df["Start Date"] = pd.to_datetime(df["Start Date"])
```

```python
df["Last Login Time"] = pd.to_datetime(df["Last Login Time"])
```

```python
#or to do it in pd.read_csv("employees.csv", parse_dates = ["Start Date", "Last Login Time"])
```

```python
df.head()
```

```python
df["Senior Management"] = df["Senior Management"].astype("bool")
```

```python
df.info()
```

```python
df["Gender"] = df["Gender"].astype("category")
```

```python
df.info()
```

```python
df["Team"] = df["Team"].astype("category")
```

```python
df.info()
```

## Filter with More than One Condition (AND &)

```python
df.head(3)
```

```python
mask1 = df["Gender"] == "Male"
mask2 = df["Team"] == "Marketing"

df[mask1 & mask2].head(3)
```

## Filter with More than One Condition (OR - |)

```python
mask1 = df["Senior Management"]
mask2 = df["Start Date"] < "1990-01-01"

df[mask1 | mask2]
```

```python
mask1 = df["First Name"] == "Robert"
mask2 = df["Team"] == "Client Services"
mask3 = df["Start Date"] > "2016-06-01"

df[(mask1 & mask2) | mask3]
```

## The isin() Method

```python
mask1 = df["Team"] == "Legal"
mask2 = df["Team"] == "Sales"
mask3 = df["Team"] == "Product"

df[mask1 | mask2 | mask3 ]
```

```python
df[df["Team"].isin(["Legal", "Product", "Sales"])]

mask = df["Team"].isin(["Legal", "Product", "Sales"])
df[mask]
```

## The .isnull() and .notnull() Methods

```python
df[df["Team"].isnull()].head(3)
```

```python
mask = df["Gender"].notnull()
df[mask]
```

## The .between() Method

```python
mask = df["Salary"].between(60000, 70000)
df[mask].sort_values("Salary")
```

```python
mask = df["Bonus %"].between(2.0 , 5.0)
df[mask]
```

```python
mask = df["Start Date"].between("1991-01-01","1992-01-01")
df[mask]
```

```python
mask = df["Last Login Time"].between("08:30 AM", "12:00 PM")
df[mask]
```

## The .duplicated() Method

```python
df.sort_values("First Name", inplace=True)
```

```python
df.head()
```

```python
df[df["First Name"].duplicated()]
```

```python
df["First Name"].duplicated(keep = "last")
```

```python
df[df["First Name"].duplicated(keep = False)]
```
