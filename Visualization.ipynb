## Bar Graphs

```python
bb = data.DataReader(name = "GOOG", data_source= "yahoo", start = "2004", end = "2016")
bb.head(3)
```

```python
def rank_perfomance(stock_price):
    if stock_price <= 200:
        return "Poor"
    elif stock_price > 200 <= 500:
        return "Satist"
    else:
        return "Stellar"
```

```python
bb["Close"].apply(rank_perfomance).value_counts().plot(kind = "barh")
```

## Pie Charts

```python
bb["Close"].mean()
```

```python
def rank(stock_price):
    if stock_price>= 313.31145301949425:
        return "Above Averege"
    else: return "Below Averege"
```

```python
bb["Close"].apply(rank).value_counts().plot(kind = "pie", legend = True)
```

## Histograms

```python
def custom(stock_price):
    return int(stock_price / 100.0) * 100
```

```python
custom(456)
```

```python
bb["High"].apply(custom).value_counts().sort_index()
```

```python
bb["High"].apply(custom).value_counts().nunique()
```

```python
bb["High"].apply(custom).plot(kind = "hist", bins = 8)
```
